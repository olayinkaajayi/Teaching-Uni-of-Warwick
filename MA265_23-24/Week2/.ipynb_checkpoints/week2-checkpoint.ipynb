## Rosenbrock Function

```python
# Define Rosenbrock Function
def rosenbrock(x, a=1, b=100):
    y = x[1]
    x = x[0]
    return (a - x)**2 + b*(y - x**2)**2
```

```python
x = np.linspace(-2, 2, 1000)
y = np.linspace(-2, 2, 1000)
X, Y = np.meshgrid(x, y)
Z = rosenbrock([X, Y])

fig = plt.figure(figsize=(16, 6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection="3d")
ax1.set_xlim([-2, 2])
ax1.set_ylim([-2, 2])
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlim([0, 2500])

# Color mesh
ax1.set_facecolor("white")
ax1.pcolormesh(X, Y, Z, cmap=cm.jet,norm=LogNorm())
ax1.scatter(1, 1, color="k")
ax1.annotate('Min', xy=(1, 1), xytext=(-0.5, 1.25),arrowprops=dict(facecolor='black', shrink=0.05))

# Surface plot
ax2.set_facecolor("white")
ax2.plot_surface(X, Y, Z, norm = LogNorm(), cmap=cm.jet, linewidth=0)
ax2.view_init(azim=65, elev=25)
ax2.scatter(1., 1., 0., color="k")
xa, ya, _ = proj3d.proj_transform(1,1,0, ax2.get_proj())
ax2.annotate('Min', xy = (xa, ya), xytext = (-20, 30), textcoords = 'offset points', ha = 'right', va = 'bottom', arrowprops=dict(facecolor='black', shrink=0.05))

plt.tight_layout()
plt.show()
```

## Gradient descent for a quadratic function
We consider a function least squares problem $$f(x,y) = \frac{1}{2}\|\mathbf{A}x-\mathbf{b}\|^2$$
where
\begin{align*}
\mathbf{A} = \begin{pmatrix}
1 & 2 \\
2 & 1 \\
-1 & 0
\end{pmatrix}
\text{ and } \mathbf{b} = \begin{pmatrix} 10 \\ -1\\ 0 \end{pmatrix}.
\end{align*}


```python
class Quadratic:
    
    def __init__(self, A, b):
        self.A = A
        self.b = b

    def f(self, x):
        return la.norm(np.dot(self.A,x)-self.b)**2/2
    
    def grad(self, x):
        return np.dot(np.transpose(self.A), np.dot(self.A, x)-self.b)
    
    def residual(self, x):
        return -self.grad(x)
    
    def step_len(self, x):
        r = self.residual(x)
        Ar = np.dot(self.A, r)
        return (la.norm(r)**2)/(la.norm(Ar)**2)
```

```python
# Alternative
# A = np.array([[1, 1], 
#              [2,-1]])
#b = np.array([[1],
#              [-1]])

A = np.array([[1,  1],
              [2,  1],
              [-1, 0]])
b = np.array([[2],
              [-5],
              [ 0]])
```

```python
# Test it
q = Quadratic(A,b)
x0 = np.array([[1],
               [1]])
q.f(x0)
```

```python
T = 10
x0 = np.array([[5],[-2]])
x = x0.copy()
evolution_xy = [[x0[0][0], x0[1][0]]]
tau = 0.1
for i in range(T):
    x = x - q.step_len(x)*q.grad(x)
    evolution_xy = np.vstack((evolution_xy, [x[0][0], x[1][0]]))

evolution_x = evolution_xy[:, 0]
evolution_y = evolution_xy[:, 1]

print('Final iterate: (', x[0][0], ', ', x[1][0], ')')
```

```python
# Compare with closed form solution
At = np.transpose(A)
x1 = np.dot(la.inv(np.dot(At,A)),np.dot(At,b))
print('Closed form solution: (', x1[0][0], ', ', x1[1][0], ')')
```

## Stochastic Gradient Descent (SGD) for a quadratic function

We consider a function least squares problem $$\frac{1}{2}\|Ax-b\|^2=\frac{1}{2}\sum_{i=1}^m (a_i^Tx-b_i)^2 $$

where we assume that $A$ is very tall ($m>>n$). Instead of computing the whole gradient, we compute the gradient in batches.

```python
class Quadratic:
    
    def __init__(self, A, b):
        self.A = A
        self.b = b

    def f(self, x):
        return la.norm(np.dot(self.A,x)-self.b)**2/2
    
    def grad(self, x, batch):
        # batch consists of the indices of which the gradient is taken
        Abatch = self.A[batch,:]
        bbatch = self.b[batch,:]
        return np.dot(np.transpose(Abatch), np.dot(Abatch, x)-bbatch)
    
    def residual(self, x):
        return -self.grad(x)
    
    def step_len(self, x):
        r = self.residual(x)
        Ar = np.dot(self.A, r)
        return (la.norm(r)**2)/(la.norm(Ar)**2)
```

## Generate data

```python
m = 1000
x = 2
A = np.linspace(0,1,m).reshape(-1,1)
b = x*A+rnd.randn(m,1)/10
```

```python
plt.plot(A, b, 'o', lw='3')
plt.show()
```

```python
# q is a quadratic function taking one x value
q = Quadratic(A, b)
```

```python
# Test it
x0 = np.array(1)
q.f(x0)
```

```python
# Stochastic gradient descent
batch_sizes = [1,10,100,m]
T = 10000
x = np.array([[0]])
lr = 1e-4 # learning rate = step length
errors = np.zeros((4,T))
```

```python
for j,b in enumerate(batch_sizes):
    for t in range(T):
        batch = random.sample(range(m), b)
        r = lr * q.grad(x, batch)
        x = x - r
        errors[j,t] = la.norm(r[0,0])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 4), layout='constrained')
for i in range(2):
    for j in range(2):
        ax[i,j].plot(np.arange(T), errors[2*i+j])
        ax[i,j].set_title(batch_sizes[2*i+j])
plt.show()
```

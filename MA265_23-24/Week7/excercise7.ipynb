```python
import numpy as np
from itertools import product
import matplotlib.pyplot as plt
```

```python
def determinant_Vandermonde(points):
    """Computes the determinant of a vandermonde matrix given the pair of points"""
    all_pairs = [tuple({a,b}) for a in points for b in points if a!=b]
    all_unique_pairs = set(all_pairs)
    pairs = [a-b for a,b in all_unique_pairs]
    det = np.product(pairs)

    return det
```

## Determinant of the Vandermonde Matrix $V$ for differently spaced values of $x_i$. s.t. $0\leq x_i \leq 5$.

Recall that

\begin{equation}
    \mathrm{det}(V) = \prod_{0\leq i,j \leq n} (x_j - x_i)
\end{equation}

```python
# See the closeness of points
spacing_list = [1, 0.5, 0.2, 0.08, 0.02]

for space in spacing_list:
    points = np.arange(0, 5, space)
    determinant = determinant_Vandermonde(points)
    plt.scatter(points,[1]*len(points))
    plt.xlabel(r"points $x_i$")
    plt.title(f"The determinant is {determinant:.4f}")
    plt.show()
```

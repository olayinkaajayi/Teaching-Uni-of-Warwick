Numerically analyse the invertibility of the Vandermonde matrix

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

# Define function which assesmbles the Vandermonde matrix 
def Vandermonde(x, d):
    V = np.vander(x, d + 1, increasing = True)
    return V

# Initialise array which contains the condition numbers of our Vandermonde matrices
kappa = []

for n in range(1, 101):
    # For 1  n  100
    
    # Generate array x of equidistant points in [1, 1] of length n
    x = np.linspace(-1, 1, n)
    
    # Generate the square Vandermonde matrix V for the array x
    V = Vandermonde(x, np.size(x) - 1)
    
    # Calculate the Condition number V wrt the 2-norm
    k = np.linalg.cond(V, 2)
    kappa.append(k)

plt.figure(figsize = (15, 10))
plt.xlabel('Size of square Vandermonde matrix V of order n', fontsize = 20)
plt.ylabel('(V)', fontsize = 20)
plt.title('Condition number (V) of V of order n for 1  n  100 wrt the 2-norm. (Logarithmic scale)', fontsize = 20)
plt.semilogy([n for n in range(1, 101)], kappa)
plt.grid()
plt.show()
```

c (i).
$\newline$
Plot the humidity measurements against their corresponding pressure measurements.

```python
# Extract data on Pressure and Humidity from csv file
Pressure = []
Humidity = []
with open('WeatherData.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Pressure.append(float(row['Pressure (millibars)']))
        Humidity.append(float(row['Humidity']))
        
# Plot the humidity measurements against their corresponding pressure measurements
plt.figure(figsize = (15, 10))
plt.xlabel('Pressure (millibars)', fontsize = 20)
plt.ylabel('Humidity', fontsize = 20)
plt.title('Humidity measurements against their corresponding pressure measurements')
plt.scatter(Pressure, Humidity)
plt.grid()
plt.show()
```

c (ii).
$\newline$
Starting from d = 0 to d = 5 identify the coefficients of the 6 polynomial models and
state their values.

```python
modelCoefficients = []

# For 0<=d<=5, calculate and display the coefficients of the polynomial model using a Vandermonde matrix
for d in range(0, 6):
    b = Humidity[:90]
    x = Pressure[:90]
    V = Vandermonde(x, d)
    a = np.linalg.lstsq(V, b, -1)[0]
    # Store coefficients of each polynomial model in modelCoefficients
    modelCoefficients.append(a)
    
    print("Coefficients of polynomial model of degree {} are: \n{}\n".format(d, a.tolist()))
```

c (iii).
$\newline$
Compute the relative forward errors of the polynomial models you identified in part
ii and then and plot them against their corresponding polynomial order.

```python
testPressure = Pressure[:90]
exactHumidity = Humidity[:90]

# Initialising the Degree and Error arrays for plotting
Degree = [d for d in range(0, 6)]
Error = []

# Loop through modelCoefficients and calculate the relative forward error for each polynomial model
for a in modelCoefficients:
    '''Evaluate the polynomial model at Pressure values from day 1 to day 90 and calculate the relative forward 
       error wrt the 2-norm'''
    approximateHumidity = np.polyval(a[::-1], testPressure)
    relativeForwardError = np.linalg.norm(approximateHumidity - exactHumidity) / np.linalg.norm(exactHumidity)
    Error.append(relativeForwardError)

print(Error)
print(Error.index(min(Error)))
# Plot the relative forward error of the polynomial models against their corresponding polynomial order 
plt.figure(figsize = (15, 10))
plt.plot(Degree, Error, '-bo')
plt.xlabel('Degree of polynomial', fontsize = 20)
plt.ylabel('Relative forward error', fontsize = 20)
plt.title('Relative forward errors of the polynomial models against their corresponding polynomial order using day 1 to day 90')
plt.grid()
plt.show()
```

c (iv).
$\newline$
Compute the relative forward errors of the polynomial models you identified in part
ii on the test data and then and plot them against their corresponding polynomial
order.

```python
# Splice Pressure and Humidity arrays to extract only the test data
testData = Pressure[90:]
exactHumidity = Humidity[90:]

# Initialising the Degree and Error arrays for plotting
Degree = [d for d in range(0, 6)]
Error = []

# Loop through modelCoefficients and calculate the relative forward error for each polynomial model
for a in modelCoefficients:
    # Evaluate the polynomial model for test data values and calculate the relative forward error wrt the 2-norm
    approximateHumidity = np.polyval(a[::-1], testData)
    relativeForwardError = np.linalg.norm(approximateHumidity - exactHumidity) / np.linalg.norm(exactHumidity)
    Error.append(relativeForwardError)

print(Error)
print(Error.index(min(Error)))
# Plot the relative forward error of the polynomial models against their corresponding polynomial order
plt.figure(figsize = (15, 10))
plt.plot(Degree, Error, '-ro')
plt.xlabel('Degree of polynomial', fontsize = 20)
plt.ylabel('Relative forward error', fontsize = 20)
plt.title('Relative forward errors of the polynomial models against their corresponding polynomial order using day 91 to day 100')
plt.grid()
plt.show()
```

c (v).

What is the optimal order of polynomial that fits your weather data. Justify your answer.
$\newline$
$Answer: $
$\newline$
The optimal order of polynomial is 2.
On the test data (day 91 to day 100) we have the minimum relative error at degree 2, even though on the training data (day 1 to day 90) the minimum relative error is at degree 5, as the degree of V continues to increase, so does the condition number; so the trade-off is not worth it and the optimal order of polynomial is 2.
